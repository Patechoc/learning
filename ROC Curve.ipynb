```python
# This tutorial assumes you've gotten to the point of having a model
# that is trained on a training set and made real-valued predictions
# on a testing set. You can use the code cells after this one on your
# problem, assuming you have the appropriate variables created.

# This is (essentially) the same code as in the "Modeling and scikit-learn" tutorial
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Series(iris['target']).apply(lambda x: iris.target_names[x])
df['is_setosa'] = df['species'] == 'setosa'
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
for f in features:
    # Add a bunch of noise, this data is too easy to separate.
    mean = df[f].mean()
    df[f] += np.random.randn(len(df[features])) * mean
X = df[features].values  # Feature matrix
y = df['is_setosa'].values  # Target label vector
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2)
print 'Splitting %d samples into %d training and %d test samples' % (
    len(X), len(X_train), len(X_test))
estimator = linear_model.LogisticRegression()
estimator.fit(X_train, y_train)
predictions = estimator.predict_proba(X_test)[:,1]
```

```python
# ROC Curve basics. It's not that hard.
# Based on http://scikit-learn.org/stable/auto_examples/plot_roc.html
from sklearn import metrics

fpr, tpr, _thresholds = metrics.roc_curve(y_test, predictions)
auc = metrics.auc(fpr, tpr)

fig, ax = plt.subplots(1)
ax.plot(fpr, tpr, label='AUC (c-statistic) = %0.2f' % auc)
ax.plot([0, 1], [0, 1], 'k--', label='Dumb classifier')
ax.set_xlabel('False Positive Rate', fontsize=14)
ax.set_ylabel('True Positive Rate', fontsize=14)
ax.set_title('Receiver Operator Characteristic', fontsize=16)
ax.legend(loc='lower right', fontsize=14)
# fig.savefig('basic_roc.png')
```

```python
# Since a ROC curve is abstract, it helps to translate into
# the direct units of interest for this classifier.

total_setosas = sum(y_test)
total_other = sum(1 - y_test)

fig, ax = plt.subplots(1)
ax.plot(fpr * total_other, tpr * total_setosas, label='Possible operating points')
ax.plot([0, total_other], [0, total_setosas], 'k--', label='Dumb classifier')
ax.set_xlabel('False positive count: Other flowers incorrectly classified as setosas', fontsize=14)
ax.set_ylabel('True positives count: setosas correctly identified', fontsize=14)
ax.set_title('Model tradeoff', fontsize=18)
ax.legend(loc='lower right', fontsize=14)
ax.tick_params(labelsize=14)
# fig.savefig('even_better_roc.png')
```
