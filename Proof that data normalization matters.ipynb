```python
# Generate two random features, one 100x larger than the other
f1 = np.random.randn(100)
fact = 100
f2 = np.random.randn(100) * fact
```

```python
# Build a classifier that is true whenever the factors add up to a certain
# value (see graph)
y =  (f1 + f2 * 1.0 / fact) > 1
# To make it an imperfect classifier, randomly switch some
for i in range(len(y)):
    if np.random.uniform() > 0.9:
        y[i] = 1 - y[i]
```

```python
def makefigure():
    fig, ax = plt.subplots(1)
    ax.set_xlabel('feature1')
    ax.set_ylabel('feature2')
    ax.set_title('Red = true')
    return fig, ax
```

```python
fig, ax = makefigure()
ax.scatter(f1,f2, color=['red' if t else 'blue' for t in y])
```

```python
X = np.array([f1, f2]).T
# Fit regular (no-normalization penalty) linear regression
from sklearn.linear_model import LinearRegression, Lasso
est = LinearRegression()
est.fit(X,y)
# Correctly assigns 100x weight to the 100x smaller feature
print est.coef_
```

```python
# Now add a regularization penalty
est = Lasso(alpha=1)
est.fit(X,y)
# One feature is pushed to zero
print est.coef_
```

```python
# Normalize the first feature to match the magnitude of the second
X = np.array([f1 * fact, f2]).T
fig, ax = makefigure()
ax.set_title('Notice that features are on the same scale now')
ax.scatter(X[:,0],X[:,1], color=['red' if t else 'blue' for t in y])
```

```python
est = Lasso(alpha=1)
est.fit(X,y)
# Much more reasonable and correct coefficients
print est.coef_
```
